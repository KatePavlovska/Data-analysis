# Task 1 
Постройте гистограммы распределения цен для каждой категории (по количеству комнат).
Какие выводы можно сделать на основе полученных гистограмм?

```python
fig = go.Figure()

for room in sorted(df.Rooms.unique()):
    room_df = df.query('Rooms == @room')
    n_bins = (room_df.Price.max() - room_df.Price.min()) // 50
    fig.add_trace(go.Histogram(x=room_df.Price, nbinsx=n_bins, opacity=0.75, name=f'room - {room}'))

fig.update_layout(height=600, width=950, xaxis=dict(range=(0, 19000), dtick=1000, title='Price'),
                 yaxis=dict(dtick=5, title='count'), barmode='overlay', title='Rooms-prices hist (bin = 100$)')
    
fig.show()
```

## Висновок: згідно отриманих результатів, можна сказати, що на ринку нерухомості є велика кількість однокімнатних квартир. Різниця між к-стю двокімнатних та трикімнатних квартир є незначною, в користь двокімнатних квартир. Є області гістограми, які "накладаються". Тобто існують варіанти, які коштують однаково, але к-сть кімнат різна, особливо така картина спостерігається для 2 та 3 - кімнатних квартир. Зі збільшенням кількості кімнат, зростає і ціна, що є очевидним фактом. На ринку нерухомості існують варіанти продажу квартир з к-стю кімнат n+1 за ціною квартири, що має n кімнат, але такі випадки поодинокі. Для більш детального аналізу потрібно розглядати ринок нерухомості враховуючи усі параметри, а не тільки ціну та к-сть кімнат в квартирі.
